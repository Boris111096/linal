```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts


%matplotlib inline
plt.rcParams["font.size"] = 15
```

### Уровень 0:
* Задание 1:  
Получить интуицию, на что влияют параметры распределений.  
Построить по 5 функций плотностей с разными параметрами для каждого распределения:  
нормальное, экспоненциальное, Стьюдента на одном графике.

```python
mu = 2.0
sigma = 0.5
norm_rv = sts.norm(loc=mu, scale=sigma)
x = np.linspace(0,4,100)
cdf = norm_rv.cdf(x)
pdf = norm_rv.pdf(x)
isf = norm_rv.isf(x)
sf = norm_rv.sf(x)
ppf = norm_rv.ppf(x)
plt.figure(figsize=(15, 15))
CDF, = plt.plot(x, cdf, label = 'cdf')
PDF, = plt.plot(x, pdf, label = 'pdf')
ISF, = plt.plot(x, isf, label = 'isf')
SF, = plt.plot(x, sf, label = 'sf')
PPF, = plt.plot(x, ppf, label = 'ppf')
plt.ylabel('$F(x)$')
plt.xlabel('$x$')
plt.legend(handles=[CDF,PDF,ISF,SF,PPF], loc='upper right')
plt.axis('scaled') 
plt.show()
```

```python
lam = 0.1
expon_rv = sts.expon(lam)
x = np.linspace(0,4,100)
cdf = expon_rv.cdf(x)
pdf = expon_rv.pdf(x)
isf = expon_rv.isf(x)
sf = expon_rv.sf(x)
ppf = expon_rv.ppf(x)
plt.figure(figsize=(15, 15))
CDF, = plt.plot(x, cdf, label = 'cdf')
PDF, = plt.plot(x, pdf, label = 'pdf')
ISF, = plt.plot(x, isf, label = 'isf')
SF, = plt.plot(x, sf, label = 'sf')
PPF, = plt.plot(x, ppf, label = 'ppf')
plt.ylabel('$F(x)$')
plt.xlabel('$x$')
plt.legend(handles=[CDF,PDF,ISF,SF,PPF], loc='upper right')
plt.axis('scaled') 
plt.show()
```
