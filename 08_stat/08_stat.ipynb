```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts


%matplotlib inline
plt.rcParams["font.size"] = 15
```

### Уровень 0:
* Задание 1:  
Получить интуицию, на что влияют параметры распределений.  
Построить по 5 функций плотностей с разными параметрами для каждого распределения:  
нормальное, экспоненциальное, Стьюдента на одном графике.

```python
x = np.linspace(0,4,100)
plt.figure(figsize=(15, 15))
for mu, sigma, leg in [[0.1,0.1,'pdf_1'],[1,0.25,'pdf_2'],[2,0.5,'pdf_3'],[3,1,'pdf_4'],[4,2,'pdf_5']]:
    norm_rv = sts.norm(loc=mu, scale=sigma)
    pdf = norm_rv.pdf(x)
    plt.plot(x,pdf,label=leg)
plt.ylabel('$F(x)$')
plt.xlabel('$x$')
plt.legend(loc='upper right')
plt.axis('scaled') 
plt.show()
```

```python
x = np.linspace(0,4,100)
plt.figure(figsize=(15, 15))
for lam, leg in [[0.1,'pdf_1'],[0.5,'pdf_2'],[1,'pdf_3'],[2,'pdf_4'],[3,'pdf_5']]:
    expon_rv = sts.expon(lam)
    pdf = expon_rv.pdf(x)
    plt.plot(x,pdf,label=leg)
plt.ylabel('$F(x)$')
plt.xlabel('$x$')
plt.legend(loc='upper right')
plt.axis('scaled') 
plt.show()
```

```python
x = np.linspace(0,4,100)
plt.figure(figsize=(15, 15))
for n, leg in [[0.1,'pdf_1'],[0.5,'pdf_2'],[1,'pdf_3'],[10,'pdf_4'],[100,'pdf_5']]:
    t_rv = sts.t(n)
    pdf = t_rv.pdf(x)
    plt.plot(x,pdf,label=leg)
plt.ylabel('$F(x)$')
plt.xlabel('$x$')
plt.legend(loc='upper right')
plt.axis('scaled') 
plt.show()
```
