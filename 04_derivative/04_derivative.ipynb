### Уровень 0:
Посчитайте значение первой и второй производной функции:  

$f(x) = x^5 + 4sin(2x) + cos(3x+3)$  

В точке x = 1.

```python
x = 1
```

#### Ответ:  
$f'(x) = 5x^4 + 8⋅cos(2x) - 3⋅sin(3x+3))$  

```python
5*x**4 + 8*np.cos(2*x) - 3*np.sin(3*x+3)
```

$f''(x) = 20x^3-16·sin(2x)-9·cos(3x+3)$  

```python
20*x**3-16*np.sin(2*x)-9*np.cos(3*x+3)
```

### Уровень 1:
Постройте график первой и второй производной функции:  

$f(x) = sin^5(2x+1)$  

На отрезке [-5; 5].

```python
# f'

X_1 = []
Y_1 = []
for i in range(-5,6,1):
    for j in range(100):
        x = i+j/100
        y = 10*(np.sin(2*x+1))**4*np.cos(2*x+1)
        X_1.append(x)
        Y_1.append(y)

plt.grid()
plt.xlim([-5, 5])
plt.plot(X_1,Y_1)
```

```python
# f''

X_2 = []
Y_2 = []
for i in range(-5,6,1):
    for j in range(100):
        x = i+j/100
        y = 80 * (np.sin(2*x + 1))**3 * (np.cos(2*x + 1))**2 - 20*(np.sin(2*x + 1))**5
        X_2.append(x)
        Y_2.append(y)

plt.grid()
plt.xlim([-5, 5])
plt.plot(X_2,Y_2)
```
